#### Pyspark Introduction:

```python
# imports:--

from pyspark.sql import SparkSession
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create spark sesseion:--

spark = SparkSession.builder.appName("practice").getOrCreate()
```

##### Spark DataFrame:

```python
data1 = [
    ("Alice", 4500), 
    ("Sarvana", 5000),
    ("Chandana", 5500),
    ("Kiran", 3400)
]

columns = ["Name", "Salary"]

df1 = spark.createDataFrame(data1, columns)
df1.show(truncate=False)
```

```python
data2 = [
    (101, "Kiran", "Reddy", "Marketing", 50000, 5.5),
    (102, "Ravi", "Kadiyala", "Sales", 45000, 4.5),
    (103, "Devansh", "Panini", "Engineering", 80000, 3.0)
]


headings = ["ID", "Firstname", "Lastname", "Department", "Salary", "Experience"]

df2 = spark.createDataFrame(data2, headings)
df2.show(truncate=False)
```

###### 2D- Numpy Array to DataFrame:-

```python
arr1 = np.array([
    [12, 23, 34],
    [45, 56, 67],
    [78, 89, 90]
])

cols = ["X", "Y", "Z"]

df3 = spark.createDataFrame(arr1, cols)
df3.show()
```

###### Dictionary of lists to Pyspark DataFrame:

```python
dict1 = {
    "Student": ["Amol", "Swara", "Akanksha", "Vijay"],
    "Marks": [45, 55, 50, 54],
    "Class": [9, 7, 8, 5]
}

p_df = pd.DataFrame(dict1)
df4 = spark.createDataFrame(p_df)
df4.show()
```

```python
dict2 = {
    "Student": ["Amol", "Swara", "Akanksha", "Vijay"],
    "Marks": [45, 55, 50, 54],
    "Class": [9, 7, 8, 5]
}

df6 = spark.createDataFrame(zip(
    dict2["Student"],
    dict2["Marks"],
    dict2["Class"]
), schema=["Student", "Marks", "Class"])

df6.show()
```

###### List of Dictionaries to Pyspark DataFrame:

```python
list1 = [
    {"Student": "Amol", "Marks": 45, "Class": 9},
    {"Student": "Swara", "Marks": 55, "Class": 7},
    {"Student": "Akanksha", "Marks": 50, "Class": 8},
    {"Student": "Vijat", "Marks": 54, "Class": 5},
]

df5 = spark.createDataFrame(list1)
df5.show()
```

1) List of Tuples to spark DataFrame
2) 2D Numpy Array to spark DataFrame
3) List of Dictionary to spark DataFrame
4) Pandas DataFrame to spark DataFrame
5) Dictionary of List to spark DataFrame
