MS SQL Server DB Table to Spark DataFrame:

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MSSQL") \
    .config("spark.jars.packages", "com.microsoft.sqlserver:mssql-jdbc:12.4.2.jre11") \
    .getOrCreate()
```

```python
jdbc_url = (
    "jdbc:sqlserver://localhost:1433;"
    "databaseName=genaiDB;"
    "encrypt=true;"
    "trustServerCertificate=true"
)

connection_props = {
    "driver": "com.microsoft.sqlserver.jdbc.SQLServerDriver",
    "user": "sparkuser",
    "password": "Spark@123"
}

df = spark.read.jdbc(url=jdbc_url, table="dbo.samples", properties=connection_props)
df.show()
df.printSchema()
```
