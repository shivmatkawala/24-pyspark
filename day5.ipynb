```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
import pandas as pd
import numpy as np
```

```python
spark = SparkSession.builder.appName("just-practice").getOrCreate()
```

```python
df_csv = spark.read.option("header", "true").csv("dataFiles/just.csv")
df_csv.show()
```

```python
# df_csv.write.json("dataFiles/just.json")

df_json = spark.read.json("dataFiles/just.json")
df_json.show()
```

```python
# df_json.write.parquet("dataFiles/just.parquet")
df_parquet = spark.read.parquet("dataFiles/just.parquet")
df_parquet.show()
```

```python
# df_parquet.write.orc("dataFiles/just.orc")
df_orc = spark.read.orc("dataFiles/just.orc/")
df_orc.show()
```
