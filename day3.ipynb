Read Data From Different types of files and create DataFrame:

```python
from pyspark.sql import SparkSession
import pandas as pd
```

```python
spark = SparkSession.builder.appName("files_to_df").getOrCreate()
```

```python
# CSV to Pyspark DataFrame:

df1 = (
    spark
    .read
    .option("header", "false")
    .option("inferSchema", "true")
    .csv("dataFiles/sample.txt")
)

# Rename columns
df1 = df1.toDF("ID", "Name", "Email", "Age", "City", "Department", "Salary", "DOB", "IsActive")

df1.show()
```

```python
p_df = pd.read_excel("dataFiles/sports_records.xlsx")

df2 = spark.createDataFrame(p_df)
df2.show()
```

```python
df3 = (
    spark.read
    .option("multiLine", "true")
    .json("dataFiles/fruits.json")
)

df3.show()
```

```python
data = {
    "id": [1, 2, 3, 4, 5],
    "name": ["Apple", "Banana", "Orange", "Mango", "Grapes"],
    "color": ["Red", "Yellow", "Orange", "Yellow", "Green"],
    "price": [120, 40, 80, 150, 90],
    "quantity": [50, 100, 75, 60, 40],
    "country": ["India", "India", "South Africa", "India", "Australia"]
}

df = pd.DataFrame(data)

df.to_parquet("dataFiles/foods.parquet", index=False)

df = spark.read.parquet("dataFiles/foods.parquet")

df.show()

```
